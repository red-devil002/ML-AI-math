```python
# import Statement
import numpy as np
import matplotlib.pyplot as plt
```

```python
# seting up the plots and time
t = np.linspace(0, 40, 1000) # Start time, finish time, n points
```

```python
# distance for robber
d_r = 2.5 * t
```

```python
# distance for sheriff
d_s = 3 * (t-5)
```

```python
# plotting
fig, ax = plt.subplots()
plt.title("A bank Robber Caught")
plt.xlabel("Time t (in mins)")
plt.ylabel("Distance d (in Km)")
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, d_r, "r")
ax.plot(t, d_s, "g")
plt.axvline(x=30, color= "blue", linestyle = "--")
_ = plt.axhline(y=75, color="blue", linestyle="--")
```

# Higher-Rank Tensors
